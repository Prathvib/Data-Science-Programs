```python
#Question)Use a car.csv dataset and perform exploratory data analysis
#using pandas and matplotlib library functions to manipulate and
#visualize the data and find insights

#subquestions are as follows

#1)Import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
#2.Load the dataset using pandas read_csv() function
#Load cars.csv dataset
car=pd.read_csv('cars.csv')
car
```

```python
#3.Display the head of the dataset using head() function
#display the head of the dataset
car.head()
```

```python
#4.Display the bottom 5 row from the dataset using tail() function
#Display the bottom 5 rows
car.tail(5)
```

```python
#5.Print summary statistical of data set using the describe() function
#Analytical summary of dataset
car.describe(include='all')
```

```python
#6.plot a histogram for all the variables
car.hist(figsize=(20,30))
```

```python
#7.Drop irrelevent columns from the datset using drop() function
#drop irrelevent column
car=car.drop(['Dimensions.Height','Dimensions.Length'],axis=1)
car.head(5)
```

```python
#8.use rename() function to rename
#Rename column name
car=car.rename(columns={"Dimensions.Width":"Width","Engine Information.Hybrid":"engine_info_hybrid"})

car.head(5)
```

```python
#9.Print the total number of duplicate row
#rows containing duplicate data
duplicate_rows_car = car[car.duplicated()]
print("Number of duplicate rows:", duplicate_rows_car.shape[0])
#count the rows before removing the data
car.count()
```

```python
#10.Remove the duplicate rows using drop_duplicates () function
#drop the duplicates
car.drop_duplicates()
car.head()
```

```python
#11.Drop the missing values from the dataset
#find the null values
print(car.isnull().sum())
```

```python
#12.Draw a correlation plot between the variables
#find the relation between the value
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
